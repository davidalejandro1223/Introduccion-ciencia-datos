```python
import pandas as pd
```

```python
cultivos = pd.read_csv('DataSet.csv')
cultivos
```

### isnull() devuelve un dataframe con la misma cantidad de columas y filas en las que verifica si cada campo esta nulo o no

```python
cultivo_nan = cultivos.isnull()
cultivo_nan
```

### Si cualquiera de los campos de 1 fila tienen un NaN se elimina

```python
cultivos_sin_nan1 = cultivos.dropna(axis=0, how='any')
cultivos_sin_nan1
```

### si toda la fila es NaN se elimina la fila

```python
cultivos_sin_nan2 = cultivos.dropna(axis=0, how='all')
cultivos_sin_nan2
```

### Si una columna presenta un NaN se elimina

```python
cultivos_sin_nan3 = cultivos.dropna(axis=1, how='any')
cultivos_sin_nan3
```

### rellena los campos que contengan NaN con el valor dado

```python
cultivos_nuevo=cultivos.fillna(value=3)
cultivos_nuevo
```

### Remplaza nos NaN con el promedio de una columna especifica

```python
cultivos_nuevo2 = cultivos.fillna(value=cultivos['ano_vigencia'].mean())
cultivos_nuevo2
```

### El metodo bfill toma el valor siguiente de donde se encuentra un NaN y lo remplaza en el NaN, el limite determina la cantidad de veces permitidas para llenar los NaN

```python
cultivos_nuevo3 = cultivos.fillna(method='bfill', limit=1)
cultivos_nuevo3['semestre_b2013_rendim_tha']
```

### El metodo bfill toma el valor anterior de donde se encuentra un NaN y lo remplaza en el NaN

```python
cultivos_nuevo4 = cultivos.fillna(method='ffill', limit=1)
cultivos_nuevo4['semestre_b2013_rendim_tha']
```

### Verifica si se cumple la condicion en cada campo de una columna

```python
cultivos['evaluacion_definitiva_semestre_a2013_area_sembr_ha']>=5
```

### Se eliminan las filas que no cumple con la condicion en la columna determinada

```python
cultivos[cultivos['evaluacion_definitiva_semestre_a2013_area_sembr_ha']>=5]
```

### Elimina las filas en las cuales la condicion de la segunda columna no se cumple y retorna los valores de la primera columna especificada

```python
cultivos['evaluacion_definitiva_semestre_a2013_area_cosech_ha'][cultivos['evaluacion_definitiva_semestre_a2013_rendim_tha']>=15]
```

### Elimina las filas que no cumplen con la compuerta logica especificada

```python
cultivos[(cultivos['evaluacion_definitiva_semestre_a2013_area_cosech_ha']>50) & (cultivos['evaluacion_definitiva_semestre_a2013_rendim_tha']<20)] 
```
