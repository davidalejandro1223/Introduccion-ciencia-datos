## EJERCICIO 1
### Crear un ejemplo de pipe con tres funciones. La primera y la tercera función deben tener argumentos diferentes al dataframe

```python
def potencia(dataframe, pot):
    return dataframe**pot

def absoluto(dataframe):
    return dataframe.abs()

def complejo(dataframe, col):
    dataframe[col] = np.array(dataframe[col], dtype='complex')
    return dataframe
```

```python
df.pipe(potencia, 3).pipe(absoluto).pipe(complejo, 'C')
```

# EJERCICIO 2
### Mediante lambda cree una fución que permita obtener el IMC a partir de un data frame que contenga el peso y la estatura de personas.

```python
arrPeso = np.random.randint(low=50, high=120, size=(10))
arrAltura = np.random.randint(low=150, high=200, size=(10))
dfIMC = pd.DataFrame(np.column_stack([arrPeso, arrAltura]), columns=['peso', 'altura'])
dfIMC
```

```python
dfIMC.apply(lambda x: x[0]/((x[1]/100)**2), axis=1)
```

# EJERCICIO 4

### Genere un dataframe donde se almacene el precio (en pesos colombianos) de algunos productos a través de los meses de un año. A partir del dataframe anterior genere un nuevo dataframe donde los precios de los productos se encuentren en dolares, euros y libras.


```python
meses = np.arange(1,13)
productos = np.array(['televisor', 'computador', 'celular', 'reloj', 'zapatos'])
precios = np.random.randint(low=100000, high=3000000, size=(12,5))
dtPrecios = pd.DataFrame(data=precios, index=meses, columns=productos)
dtPrecios
```

```python
def libras(dataframe):
    return dataframe/4238
def dolares(dataframe):
    return dataframe/3444
def euros(dataframe):
    return dataframe/3731    
```

```python
dtPrecios.transform([libras, dolares, euros])
```
