```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfTitanic = pd.read_csv('titanic.csv')
dfTitanic
```

```python
menSurvivors = dfTitanic['Survived'][dfTitanic['Sex']=='male'].sum()
womenSurvivors = dfTitanic['Survived'][dfTitanic['Sex']=='female'].sum()
nonSurvivingMen = dfTitanic['Survived'][dfTitanic['Sex']=='male'].count() - menSurvivors
nonSurvivingWomen = dfTitanic['Survived'][dfTitanic['Sex']=='female'].count() - womenSurvivors

width = 0.4
x_pos = np.arange(2)

fig, ax = plt.subplots()

survivorsValues = np.array([menSurvivors, womenSurvivors])
nonSurvivorsValues = np.array([nonSurvivingMen, nonSurvivingWomen])

survivorsBar = ax.bar(x_pos - width/2, survivorsValues, width, label='sobrevivientes')
nonSurvivorsBar = ax.bar(x_pos + width/2, nonSurvivorsValues, width, label='no sobrevivientes')
ax.set_xticks(x_pos)
ax.set_xticklabels(('Hombres', 'Mujeres'))
ax.set_ylabel('Numero de personas')
ax.set_title('Victimas del accidente del Titanic')
ax.legend()
fig.show


print('El numero de hombres que sobrevivieron es {}'.format(menSurvivors))
print('El numero de hombres que NO sobrevivieron es {}'.format(nonSurvivingMen))
print('El numero de mujeres que sobrevivieron es {}'.format(womenSurvivors))
print('El numero de mujeres que NO sobrevivieron es {}'.format(nonSurvivingWomen))
```

```python
class3byage = dfTitanic[dfTitanic['Pclass']==1]
fare = np.array(dfTitanic['Fare'])
age = np.array(dfTitanic['Age'])

plt.scatter(age, fare, alpha=0.5)
plt.xlabel('Edad de los usuarios')
plt.ylabel('Precio del tiquete')
plt.title('Relacion de la edad con el precio pagado en primera clase')
plt.show()
```

#### Line Plot

```python
dfAir = pd.read_csv('mole_fraction_of_carbon_dioxide_in_air.csv')
dfAir
```

```python
dataM = dfAir['data_mean_global']
year = dfAir['year']
plt.plot(year, dataM)
plt.xlim(1800, 2015)
plt.ylim(250, 450);
plt.xlabel('Años')
plt.ylabel('Fracción de Dioxido de Carbono')
plt.title('Fracción de Dioxido de Carbono en el aire desde el año 1800 hasta el 2014\n')
plt.show()
```

#### Hist
##### Data set de Titanic
##### Se realizan dos implementaciones: por pyplot y por pandas. Se obtiene el mismo resultado

```python
age = np.array(dfTitanic['Age'])
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
plt.hist(age, bins=80, color='#86bf91', zorder=2, rwidth=0.9)
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Frecuencias\n')
fig.show()
```

```python
fig, axes = plt.subplots(figsize=(20, 10))
dfTitanic.hist(column = 'Age', bins=80, grid=False, figsize=(20,10), color='#86bf91', zorder=2, rwidth=0.9, ax=axes)

plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Frecuencias\n')
fig.show()
```

#### Pie
##### Data set de Titanic

```python
vc = dfTitanic['Pclass'].value_counts()
df = vc.rename_axis('Clase').reset_index(name='Cantidad')
df
```

```python
plt.pie(df['Cantidad'], labels=df['Clase'], shadow=False, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
plt.tight_layout()
plt.title('Cantidad de usuarios por clase')
plt.show()
```
